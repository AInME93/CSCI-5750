```python
#5 Dividing the dataset into separate training and test dataset (Week2 practice Step 5a: Create a Test Set)

# Requirement: Correctly set train and test data. In this work, you should randomly divide the dataset into a training dataset and a test 
# dataset where the training dataset will contain 80% of the samples and the test dataset will contain 20% samples, respectively. 

import numpy as np
np.random.seed(10)

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

train_set, test_set = split_train_test(housing, 0.2)

# train_set.info()
# train_set[train_set.isna().any(axis=1)]
```

```python
#6 Prepare training features and labels (Week 2 practice Step 8: Prepare the Data for Machine Learning Algorithms)
# Since we consider this as a supervised learning problem, we need to prepare the input features (X) and output labels (Y) 
# separately for the algorithm to learn the mapping function.
# 
# (1) Write codes to save features and output labels in two different variables 'train_features' and 'train_labels'.

train_labels = train_set["Target_out"].copy() # get labels for output label Y
train_features = train_set.drop("Target_out", axis=1) # drop labels to get features X for training set

# (2) Print out the data frame in variable 'train_features' and print output variable 'train_labels'.

print("Train Labels: ","\n", train_labels,"\n\n" ,"Train features: ", "\n",train_features)

```

```python
train_features.boxplot(figsize=(16,8))
```

```python
#7 Perform Feature Scaling - Normalization or Standardization on the dataset
# (1) perform Min-Max normalization (or standardization) to rescale the features so that they'll have similar scales

# Note 1: Standardization: The mean and standard deviation must be derived from the training data only, and then use the estimated mean & std 
# to rescale the training and test data.

# from sklearn.pipeline import Pipeline
# from sklearn.preprocessing import StandardScaler
# num_pipeline = Pipeline([('imputer', SimpleImputer(strategy="median")), ('attribs_adder', CombinedAttributesAdder()),('std_scaler', StandardScaler()),])
# housing_num_tr = num_pipeline.fit_transform(housing_num)


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() ## define the transformer
scaler.fit(train_features) ## call .fit() method to calculate the min and max value for each column in dataset
train_features_scaled = scaler.transform(train_features)

print("Min of each column: ",scaler.data_min_)
print("Max of each column: ",scaler.data_max_)


```

```python
# (2) After performing the feature scaling, visualize the data using boxplot again to examine the distribution of columns
# train_set_scaled_df = pd.DataFrame(train_set_scaled)
# train_set_scaled_df.describe()


pd.DataFrame(train_features_scaled).boxplot(figsize=(16,8))
```

```python
# Task 8: (4 points) Build several regression models. (Week practice Step 10a: Training ML model on the Training Set)

# Requirement:

# The example codes will be provided. You need to fit each machine learning model on the training data and make predictions on the test data.

# (1) Linear regression

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
lin_reg = LinearRegression() 
lin_reg.fit(train_features_scaled, train_labels)
```

```python
# (2) DecisionRegressor: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0) 
regressor.fit(train_features_scaled, train_labels)
```

```python
# Task 9: (8 points) Evaluate regression models on the training data set. (Refer to Week practice Step 10b, c)

# Requirement:

# (1) Make a prediction using the two pre-trained ML models on the Training Set

training_predictions_linear = lin_reg.predict(train_features_scaled)
training_predictions_dtr = regressor.predict(train_features_scaled)

training_predictions_linear
 
```

```python

# (2) Make a scatter plot using both variable 'training_predictions' and variable 'train_label' from two ML models

## plot scatter plot 
import matplotlib.pyplot as plt

plt.figure(figsize=(30,30))
plt.subplot(331)

plt.scatter(training_predictions_linear, train_labels)
plt.xlabel("training_predictions", fontsize=20)
plt.ylabel("train_label", fontsize=20)
plt.title("Linear Regression Scatter plot", fontsize=20)

plt.subplot(332)

plt.scatter(training_predictions_dtr, train_labels)
plt.xlabel("training_predictions", fontsize=20)
plt.ylabel("train_label", fontsize=20)
plt.title("DecisionTreeRegressor Regression Scatter plot", fontsize=20)

plt.show()
```

```python
# (3) Report the RMSE of two models in your training dataset. 

from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(train_labels, training_predictions_linear)
lin_rmse = np.sqrt(lin_mse)


dtr_mse = mean_squared_error(train_labels, training_predictions_dtr)
dtr_rmse = np.sqrt(dtr_mse)

print("Root mean square for linear regression: ",lin_rmse,",","Root mean square for decision tree regressor: ", dtr_rmse)
```

```python
# Task 10: (8 points) Evaluate regression models on the test set. (Refer to Week practice Step 12)

# Requirement:

# (1) Provide codes to apply the same data processing pipeline above on the test set

test_labels = test_set["Target_out"].copy() # get labels for output label Y
test_features = test_set.drop("Target_out", axis=1) # drop labels to get features X for training set

test_features_scaled = scaler.transform(test_features)


```

```python
# (2) Make predictions using the two regression models for test instances

test_predictions_linear = lin_reg.predict(test_features_scaled)
test_predictions_dtr = regressor.predict(test_features_scaled)
```

```python
# (3) Make a scatter plot for both variable 'test_predictions' and variable 'test_label'
plt.figure(figsize=(30,30))
plt.subplot(331)

plt.scatter(test_predictions_linear, test_labels)
plt.xlabel("test_predictions", fontsize=20)
plt.ylabel("test_label", fontsize=20)
plt.title("Linear Regression Scatter plot", fontsize=20)

plt.subplot(332)

plt.scatter(test_predictions_dtr, test_labels)
plt.xlabel("test_predictions", fontsize=20)
plt.ylabel("test_label", fontsize=20)
plt.title("DecisionTreeRegressor Regression Scatter plot", fontsize=20)

plt.show()
```

```python
# (4) Report the RMSE of two models in your test dataset. 
lin_mse = mean_squared_error(test_labels, test_predictions_linear)
lin_rmse = np.sqrt(lin_mse)


dtr_mse = mean_squared_error(test_labels, test_predictions_dtr)
dtr_rmse = np.sqrt(dtr_mse)

print("Root mean square for linear regression: ",lin_rmse,",","Root mean square for decision tree regressor: ", dtr_rmse)
```
