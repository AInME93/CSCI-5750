```python
def encode(string):
    encoded_list = list()
    string_list = list(string)
    
    for x in string_list:
        if x.upper() == 'A':
            encoded_list.append([1,0,0,0])
        elif x.upper() == 'T':
            encoded_list.append([0,1,0,0])
        elif x.upper() == 'G':
            encoded_list.append([0,0,1,0])
        elif x.upper() == 'C':
            encoded_list.append([0,0,0,1])
        else:
            encoded_list.append([0,0,0,0])
    
    return encoded_list
```
